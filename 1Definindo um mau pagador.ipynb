#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
demo_semna = demo.drop("sexo", axis=1)
````

````python
demo_semna["posse_de_veiculo"].unique()
demo_semna["posse_de_imovel"].unique()
demo_semna["tipo_renda"].unique()
demo_semna["educacao"].unique()
demo_semna["estado_civil"].unique()
demo_semna["tipo_residencia"].unique()
````

````python
demo_dummies = pd.get_dummies(demo_semna, columns=["posse_de_veiculo","posse_de_imovel","tipo_renda","educacao","estado_civil","tipo_residencia"], drop_first=True)
````

````python
df_sna = pd.DataFrame(demo_dummies)
df_sna
````
